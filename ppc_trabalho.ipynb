# Carros e caminhões

```python
import threading
import random
import time

class Caminhoes(threading.Thread):
    def __init__(self, ponte, sentido):
        threading.Thread.__init__(self)
        self.tipo = "caminhao"
        self.ponte = ponte
        self.sentido = sentido

    def run(self):
        time.sleep(random.uniform(1, 3))  # Tempo de chegada dos caminhões
        self.ponte.atravessar(self)

class Carros(threading.Thread):
    def __init__(self, ponte, sentido):
        threading.Thread.__init__(self)
        self.tipo = "carro"
        self.ponte = ponte
        self.sentido = sentido

    def run(self):
        time.sleep(random.uniform(1, 3))  # Tempo de chegada dos carros
        self.ponte.atravessar(self)

class Pontes:
    def __init__(self):
        self.mutex = threading.Lock()
        self.carros_count = 0
        self.carros_atravessados_AtoB = 0
        self.carros_atravessados_BtoA = 0
        self.caminhoes_atravessados_AtoB = 0
        self.caminhoes_atravessados_BtoA = 0 
        self.tempo_espera = []
        self.max_carros = 5
        self.travessia = 0

    def atravessar(self, veiculo):
        while True:
            start_espera = time.time()
            self.mutex.acquire()
            if veiculo.tipo == "caminhao":
                if veiculo.sentido == "AtoB" and self.travessia == 0:
                    self.caminhoes_atravessados_AtoB += 1
                    time.sleep(10)
                    if self.carros_atravessados_BtoA != 50 or self.caminhoes_atravessados_BtoA != 5:
                        self.carros_count = 0
                        self.travessia = 1
                    print(f"O caminhao {self.caminhoes_atravessados_AtoB} AtoB passou a ponte")
                    fim_espera = time.time()
                    self.tempo_espera.append(fim_espera - start_espera)
                    self.mutex.release()
                    break
                elif veiculo.sentido == "BtoA" and self.travessia == 1:
                    self.caminhoes_atravessados_BtoA += 1
                    time.sleep(10)
                    if self.carros_atravessados_AtoB != 50 or self.caminhoes_atravessados_AtoB != 5:
                        self.carros_count = 0
                        self.travessia = 0
                    print(f"O caminhao {self.caminhoes_atravessados_BtoA} BtoA passou a ponte")
                    fim_espera = time.time()
                    self.tempo_espera.append(fim_espera - start_espera)
                    self.mutex.release()
                    break
                else:
                    self.mutex.release()
            if veiculo.tipo == "carro":
                if veiculo.sentido == "AtoB" and self.travessia == 0:
                    if self.carros_count == 0:
                        self.carros_count += 1
                        self.carros_atravessados_AtoB += 1
                        time.sleep(5)
                        if self.carros_atravessados_AtoB == 50 and self.caminhoes_atravessados_AtoB == 5:
                            self.carros_count = 0
                            self.travessia = 1
                        print(f"O carro {self.carros_atravessados_AtoB} AtoB passou a ponte com count 0")
                        fim_espera = time.time()
                        self.tempo_espera.append(fim_espera - start_espera)
                        self.mutex.release()
                        break
                    elif self.carros_count == 5 and (self.carros_atravessados_BtoA != 50 or self.caminhoes_atravessados_BtoA != 5):
                        self.carros_count = 0
                        self.travessia = 1
                        self.mutex.release()
                    else:
                        decisao = random.randint(0,1)
                        if decisao == 0:
                            self.carros_count += 1
                            self.carros_atravessados_AtoB += 1
                            time.sleep(1)
                            if self.carros_atravessados_AtoB == 50 and self.caminhoes_atravessados_AtoB == 5:
                                self.carros_count = 0
                                self.travessia = 1
                            print(f"O carro {self.carros_atravessados_AtoB} AtoB passou a ponte com count diferente de 0")
                            fim_espera = time.time()
                            self.tempo_espera.append(fim_espera - start_espera)
                            self.mutex.release()
                            break
                        else:
                            if self.carros_atravessados_BtoA != 50 or self.caminhoes_atravessados_BtoA != 5:
                                self.carros_count = 0
                                self.travessia = 1
                                self.mutex.release()
                            else:
                                self.carros_count += 1
                                self.carros_atravessados_AtoB += 1
                                time.sleep(1)
                                print(f"O carro {self.carros_atravessados_AtoB} AtoB passou a ponte com count diferente de 0")
                                fim_espera = time.time()
                                self.tempo_espera.append(fim_espera - start_espera)
                                self.mutex.release()
                                break

                elif veiculo.sentido == "BtoA" and self.travessia == 1:
                    if self.carros_count == 0:
                        self.carros_count += 1
                        self.carros_atravessados_BtoA += 1
                        time.sleep(5)
                        if self.carros_atravessados_BtoA == 50 and self.caminhoes_atravessados_BtoA == 5:
                            self.carros_count = 0
                            self.travessia = 0
                        print(f"O carro {self.carros_atravessados_BtoA} BtoA passou a ponte com count 0")
                        fim_espera = time.time()
                        self.tempo_espera.append(fim_espera - start_espera)
                        self.mutex.release()
                        break
                    elif self.carros_count == 5 and (self.carros_atravessados_AtoB != 50 or self.caminhoes_atravessados_AtoB != 5):
                        self.carros_count = 0
                        self.travessia = 0
                        self.mutex.release()
                    else:
                        decisao = random.randint(0,1)
                        if decisao == 0:
                            self.carros_count += 1
                            self.carros_atravessados_BtoA += 1
                            time.sleep(1)
                            if self.carros_atravessados_BtoA == 50 and self.caminhoes_atravessados_BtoA == 5:
                                self.carros_count = 0
                                self.travessia = 0
                            print(f"O carro {self.carros_atravessados_BtoA} BtoA passou a ponte com count diferente de 0")
                            fim_espera = time.time()
                            self.tempo_espera.append(fim_espera - start_espera)
                            self.mutex.release()
                            break
                        else:
                            if self.carros_atravessados_AtoB != 50 or self.caminhoes_atravessados_AtoB != 5:
                                self.carros_count = 0
                                self.travessia = 0
                                self.mutex.release()
                            else:
                                self.carros_count += 1
                                self.carros_atravessados_BtoA += 1
                                time.sleep(1)
                                print(f"O carro {self.carros_atravessados_BtoA} BtoA passou a ponte com count diferente de 0")
                                fim_espera = time.time()
                                self.tempo_espera.append(fim_espera - start_espera)
                                self.mutex.release()
                                break
                else:
                    self.mutex.release()

def main():
    ponte = Pontes()
    veiculos = []
    for i in range(5):
        for _ in range(10):
            car1 = Carros(ponte, "AtoB")
            car2 = Carros(ponte, "BtoA")
            veiculos.append(car1)
            veiculos.append(car2)
        caminhao1 = Caminhoes(ponte, "AtoB")
        caminhao2 = Caminhoes(ponte, "BtoA")
        veiculos.append(caminhao1)
        veiculos.append(caminhao2)

    start_time = time.time()

    for veiculo in veiculos:
        veiculo.start()

    for veiculo in veiculos:
        veiculo.join()

    end_time = time.time()

    print("Quantidade de veículos que cruzaram de A para B:", ponte.carros_atravessados_AtoB + ponte.caminhoes_atravessados_AtoB)
    print("Quantidade de veículos que cruzaram de B para A:", ponte.carros_atravessados_BtoA + ponte.caminhoes_atravessados_BtoA)
    print("Tempo mínimo de espera na fila:", min(ponte.tempo_espera), "segundos")
    print("Tempo máximo de espera na fila:", max(ponte.tempo_espera), "segundos")
    print("Tempo médio de espera na fila:", sum(ponte.tempo_espera) / len(ponte.tempo_espera), "segundos")
    print("Tempo de utilização da ponte:", end_time - start_time, "segundos")

if __name__ == "__main__":
    main()
```

# Apenas carros

```python
import threading
import random
import time

class Carros(threading.Thread):
    def __init__(self, ponte, sentido):
        threading.Thread.__init__(self)
        self.ponte = ponte
        self.sentido = sentido

    def run(self):
        time.sleep(random.uniform(1, 3))  # Tempo de chegada dos carros
        self.ponte.atravessar(self)

class Pontes:
    def __init__(self):
        self.mutex = threading.Lock()
        self.carros_count = 0
        self.carros_atravessados_AtoB = 0
        self.carros_atravessados_BtoA = 0
        self.tempo_espera = []
        self.max_carros = 5
        self.travessia = 0

    def atravessar(self, carro):
        while True:
            start_espera = time.time()
            self.mutex.acquire()
            if carro.sentido == "AtoB" and self.travessia == 0:
                if self.carros_count == 0:
                    self.carros_count += 1
                    self.carros_atravessados_AtoB += 1
                    time.sleep(5)          #Tempo de travessia
                    #Transição de sentido
                    if self.carros_atravessados_AtoB == 50:
                        self.carros_count = 0
                        self.travessia = 1
                    print(f"O carro {self.carros_atravessados_AtoB} AtoB passou a ponte com count 0")
                    fim_espera = time.time()
                    self.tempo_espera.append(fim_espera - start_espera)
                    self.mutex.release()
                    break
                #Transição de sentido
                elif self.carros_count == 5 and self.carros_atravessados_BtoA != 50:
                    self.carros_count = 0
                    self.travessia = 1
                    self.mutex.release()
                else:
                    decisao = random.randint(0,1)
                    if decisao == 0:
                        self.carros_count += 1
                        self.carros_atravessados_AtoB += 1
                        time.sleep(1)       #Tempo de travessia
                        #Transição de sentido
                        if self.carros_atravessados_AtoB == 50:
                            self.carros_count = 0
                            self.travessia = 1
                        print(f"O carro {self.carros_atravessados_AtoB} AtoB passou a ponte com count diferente de 0")
                        fim_espera = time.time()
                        self.tempo_espera.append(fim_espera - start_espera)
                        self.mutex.release()
                        break
                    else:
                        #Transição de sentido
                        if self.carros_atravessados_BtoA != 50:
                            self.carros_count = 0
                            self.travessia = 1
                            self.mutex.release()
                        else:
                            self.carros_count += 1
                            self.carros_atravessados_AtoB += 1
                            time.sleep(1)   #Tempo de travessia
                            print(f"O carro {self.carros_atravessados_AtoB} AtoB passou a ponte com count diferente de 0")
                            fim_espera = time.time()
                            self.tempo_espera.append(fim_espera - start_espera)
                            self.mutex.release()
                            break

            elif carro.sentido == "BtoA" and self.travessia == 1:
                if self.carros_count == 0:
                    self.carros_count += 1
                    self.carros_atravessados_BtoA += 1
                    time.sleep(5)           #Tempo de travessia
                    #Transição de sentido
                    if self.carros_atravessados_BtoA == 50:
                        self.carros_count = 0
                        self.travessia = 0
                    print(f"O carro {self.carros_atravessados_BtoA} BtoA passou a ponte com count 0")
                    fim_espera = time.time()
                    self.tempo_espera.append(fim_espera - start_espera)
                    self.mutex.release()
                    break
                #Transição de sentido
                elif self.carros_count == 5 and self.carros_atravessados_AtoB != 50:
                    self.carros_count = 0
                    self.travessia = 0
                    self.mutex.release()
                else:
                    decisao = random.randint(0,1)
                    if decisao == 0:
                        self.carros_count += 1
                        self.carros_atravessados_BtoA += 1
                        time.sleep(1)       #Tempo de travessia
                        #Transição de sentido
                        if self.carros_atravessados_BtoA == 50:
                            self.carros_count = 0
                            self.travessia = 0
                        print(f"O carro {self.carros_atravessados_BtoA} BtoA passou a ponte com count diferente de 0")
                        fim_espera = time.time()
                        self.tempo_espera.append(fim_espera - start_espera)
                        self.mutex.release()
                        break
                    else:
                        #Transição de sentido
                        if self.carros_atravessados_AtoB != 50:
                            self.carros_count = 0
                            self.travessia = 0
                            self.mutex.release()
                        else:
                            self.carros_count += 1
                            self.carros_atravessados_BtoA += 1
                            time.sleep(1)  #Tempo de travessia
                            print(f"O carro {self.carros_atravessados_BtoA} BtoA passou a ponte com count diferente de 0")
                            fim_espera = time.time()
                            self.tempo_espera.append(fim_espera - start_espera)
                            self.mutex.release()
                            break
            else:
                self.mutex.release()

def main():
    ponte = Pontes()
    carros = []
    #Criação dos carros
    for _ in range(50):
        car1 = Carros(ponte, "AtoB")
        car2 = Carros(ponte, "BtoA")
        carros.append(car1)
        carros.append(car2)
    #Início do uso da ponte
    start_time = time.time()

    #Início das travessias
    for car in carros:
        car.start()

    #Sincronização das passagens
    for car in carros:
        car.join()
    #Fim do uso da ponte
    end_time = time.time()

    #prints dos resultados
    print("Quantidade de veículos que cruzaram de A para B:", ponte.carros_atravessados_AtoB)
    print("Quantidade de veículos que cruzaram de B para A:", ponte.carros_atravessados_BtoA)
    print("Tempo mínimo de espera na fila:", min(ponte.tempo_espera), "segundos")
    print("Tempo máximo de espera na fila:", max(ponte.tempo_espera), "segundos")
    print("Tempo médio de espera na fila:", sum(ponte.tempo_espera) / len(ponte.tempo_espera), "segundos")
    print("Tempo de utilização da ponte:", end_time - start_time, "segundos")

if __name__ == "__main__":
    main()
```
